```python
import cv2 as cv
import pytesseract
import numpy as np
from matplotlib import pyplot as plt
```

```python
img_path='Pics/1.png'
```

```python
img=cv.imread(img_path)
```

```python
def plotImages(images_arr):
    dpi=80
    img1=plt.imread(images_arr)
    height,width,depth=img1.shape
    fig_size=width/float(dpi),height/float(dpi)
    fig=plt.figure(figsize=fig_size)
    # fig, axes = plt.subplots(1, 1, figsize=fig_size)
    ax=fig.add_axes([0,0,1,1])
    ax.imshow(img1,cmap='gray')
    # plt.imshow(img1,cmap='gray')
    # plt.tight_layout()
    plt.show()
```

```python
plotImages(img_path)
```

```python
g_img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imwrite('temp_pics/gray_img.png',g_img)
```

```python
plotImages("temp_pics/gray_img.png")
```
